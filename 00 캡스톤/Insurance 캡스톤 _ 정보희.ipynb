```python
def make_category_summary(df, col):
    """
    ë²”ì£¼í˜• ë³€ìˆ˜(col)ì— ëŒ€í•´
    count / ratio / percent ìš”ì•½ ë°ì´í„°í”„ë ˆì„ ìƒì„±
    """
    value_list = df[col].unique()
    counts = []

    for v in value_list:
        counts.append(len(df.query(f"{col} == @v")))

    summary_df = pd.DataFrame({
        col: value_list,
        'count': counts
    })

    total = summary_df['count'].sum()
    summary_df['ratio'] = summary_df['count'] / total
    summary_df['percent'] = (summary_df['ratio'] * 100).round(2)

    return summary_df
```

```python
sex_df      = make_category_summary(df_plot, 'sex')
children_df = make_category_summary(df_plot, 'children')
smoker_df   = make_category_summary(df_plot, 'smoker')
region_df   = make_category_summary(df_plot, 'region')
```

```python
import numpy as np
import pandas as pd
from scipy.stats import normaltest, bartlett, levene

data = ttest_df[['smoker_yes','smoker_no']]

normal_dist = True
report = []

series_dict = {}  # dropnaëœ ì‹œë¦¬ì¦ˆ ì €ì¥

for c in data.columns:
    srs = data[c].dropna()
    series_dict[c] = srs

    # normaltestëŠ” í‘œë³¸ì´ ë„ˆë¬´ ì ìœ¼ë©´ ì˜ë¯¸ ì—†ì„ ìˆ˜ ìˆì–´ìš”(ë³´í†µ n>=8 ê¶Œì¥)
    s, p = normaltest(srs)

    normalize = p > 0.05
    report.append({"field": c, "statistic": s, "p-value": p, "result": normalize})
    normal_dist = normal_dist and normalize

# ë“±ë¶„ì‚° ê²€ì • (ë‘ ì§‘ë‹¨)
x = series_dict['smoker_yes']
y = series_dict['smoker_no']

if normal_dist:
    n = "Bartlett"
    s, p = bartlett(x, y)
else:
    n = "Levene"
    s, p = levene(x, y, center='median')

report.append({"field": n, "statistic": s, "p-value": p, "result": p > 0.05})

report_df = pd.DataFrame(report).set_index('field')
report_df
```

> ì •ê·œì„±ê³¼ ë“±ë¶„ì‚°ì„±ì„ ì¶©ì¡±í•˜ì§€ ì•Šì•„ Welch's t-test ë¥¼ ì§„í–‰í•´ì•¼í•œë‹¤

| êµ¬ë¶„        | ê¸°í˜¸      | ë‚´ìš©                                 |
| --------- | ------- | ---------------------------------- |
| ê·¸ë£¹ 1 í‰ê·    | Î¼â‚      | í¡ì—°ì(smoker = yes)ì˜ ì˜ë£Œë¹„(charges) í‰ê·  |
| ê·¸ë£¹ 2 í‰ê·    | Î¼â‚‚      | ë¹„í¡ì—°ì(smoker = no)ì˜ ì˜ë£Œë¹„(charges) í‰ê·  |
| ê·€ë¬´ê°€ì„¤ (Hâ‚€) | Î¼â‚ = Î¼â‚‚ | í¡ì—°ìì™€ ë¹„í¡ì—°ìì˜ ì˜ë£Œë¹„ í‰ê· ì—ëŠ” ì°¨ì´ê°€ ì—†ë‹¤         |
| ëŒ€ë¦½ê°€ì„¤ (Hâ‚) | Î¼â‚ â‰  Î¼â‚‚ | í¡ì—°ìì™€ ë¹„í¡ì—°ìì˜ ì˜ë£Œë¹„ í‰ê· ì—ëŠ” ì°¨ì´ê°€ ìˆë‹¤         |



```python
from scipy.stats import ttest_ind

# ë‘ ê·¸ë£¹ ë°ì´í„° (NaN ì œê±°)
group1 = ttest_df['smoker_yes'].dropna()   # í¡ì—°ì
group2 = ttest_df['smoker_no'].dropna()    # ë¹„í¡ì—°ì

# Welch's t-test (equal_var=False)
s, p = ttest_ind(
    group1,
    group2,
    alternative='two-sided',
    equal_var=False
)

print(
    "statistic: {0:.3f}, p-value: {1:.3e}, {2}ê°€ì„¤ ì±„íƒ"
    .format(s, p, "ëŒ€ë¦½" if p <= 0.05 else "ê·€ë¬´")
)
```

> Welchâ€™s t-test ê²°ê³¼, í¡ì—°ìì™€ ë¹„í¡ì—°ìì˜ í‰ê·  ì˜ë£Œë¹„ ì°¨ì´ëŠ” í†µê³„ì ìœ¼ë¡œ ë§¤ìš° ìœ ì˜í•˜ì˜€ë‹¤
(t = 32.75, p < 0.001).
ë”°ë¼ì„œ ê·€ë¬´ê°€ì„¤ì„ ê¸°ê°í•˜ê³ , í¡ì—° ì—¬ë¶€ì— ë”°ë¼ ì˜ë£Œë¹„ í‰ê· ì— ì°¨ì´ê°€ ìˆë‹¤ëŠ” ëŒ€ë¦½ê°€ì„¤ì„ ì±„íƒí•œë‹¤.

- ë¹„í¡ì—°ìì˜ ì˜ë£Œë¹„ê°€ í¡ì—°ìì˜ ì˜ë£Œë¹„ë³´ë‹¤ ì‘ì€ì§€ ê²€ì •
(= í¡ì—°ìê°€ ë” ë¹„ì‹¼ì§€ ê²€ì •)

| êµ¬ë¶„        | ìˆ˜ì‹             | ë‚´ìš©                         |
| --------- | -------------- | -------------------------- |
| ê·€ë¬´ê°€ì„¤ (Hâ‚€) | Î¼(no) â‰¥ Î¼(yes) | ë¹„í¡ì—°ìì˜ í‰ê·  ì˜ë£Œë¹„ëŠ” í¡ì—°ìë³´ë‹¤ ì‘ì§€ ì•Šë‹¤  |
| ëŒ€ë¦½ê°€ì„¤ (Hâ‚) | Î¼(no) < Î¼(yes) | **ë¹„í¡ì—°ìì˜ í‰ê·  ì˜ë£Œë¹„ëŠ” í¡ì—°ìë³´ë‹¤ ì‘ë‹¤** |


```python
from scipy.stats import ttest_ind

non_smoker = ttest_df['smoker_no'].dropna()
smoker = ttest_df['smoker_yes'].dropna()

s, p = ttest_ind(
    non_smoker,
    smoker,
    alternative='less',
    equal_var=False    # Welch's t-test
)

print(
    "statistic: {0:.3f}, p-value: {1:.3e}, {2}ê°€ì„¤ ì±„íƒ"
    .format(s, p, "ëŒ€ë¦½" if p <= 0.05 else "ê·€ë¬´")
)
```

```python
from scipy.stats import normaltest, bartlett, levene
from pandas import DataFrame

equal_var_fields = []
normal_dist = True
report = []

for c in data.columns:
    series = data[c].dropna()   # â­ ë°˜ë“œì‹œ NaN ì œê±°

    s, p = normaltest(series)

    normalize = p > 0.05
    report.append({
        "field": c,
        "statistic": s,
        "p-value": p,
        "result": normalize
    })

    normal_dist = normal_dist and normalize


# ë³€ìˆ˜ê°€ ë‘ ê°œ ì´ìƒì¸ ê²½ìš° ë“±ë¶„ì‚°ì„± í™•ì¸
if len(data.columns) > 1:
    if normal_dist:
        n = "Bartlett"               # ì •ê·œì„± ì¶©ì¡± ì‹œ
        s, p = bartlett(*[data[c].dropna() for c in data.columns])
    else:
        n = "Levene"                 # ì •ê·œì„± ë¯¸ì¶©ì¡± ì‹œ
        s, p = levene(*[data[c].dropna() for c in data.columns], center="median")

    report.append({
        "field": n,
        "statistic": s,
        "p-value": p,
        "result": p > 0.05
    })

report_df = DataFrame(report).set_index("field")
report_df
```

```python
# ì´ìƒì¹˜ì™€ ì™œë„ ê²€ì •
data = df_corr[['age', 'bmi', 'children', 'log_charges']]
results = {}

for col in data.columns:
    temp = data[col].dropna()

    # ì™œë„ ê³„ì‚°
    skew_original = temp.skew()

    # Z-score ê¸°ë°˜ ì´ìƒì¹˜ íƒì§€ (|z| > 3)
    z_scores = zscore(temp)
    outlier_count = int(np.sum(np.abs(z_scores) > 3))

    # ë¡œê·¸ ë³€í™˜ í›„ ì™œë„ ê³„ì‚° (ìŒìˆ˜ ëŒ€ë¹„ +1)
    data_log = np.log1p(temp - temp.min() + 1)
    skew_log = data_log.skew()

    # ë¡œê·¸ ë³€í™˜ í•„ìš” ì—¬ë¶€ íŒë‹¨
    need_log = (abs(skew_original) > 1) and (abs(skew_log) < abs(skew_original))

    results[col] = {
        'original_skew': skew_original,
        'log_skew': skew_log,
        'outliers(|z|>3)': outlier_count,
        'log_recommended': need_log,
    }

# ê²°ê³¼ dataframe ë³€í™˜
results_df = pd.DataFrame(results).T
display(results_df)

# ì´ìƒì¹˜ ê²€ì • ê²°ê³¼
outlier_flag = len(results_df[results_df['outliers(|z|>3)'] > 0]) > 0


if outlier_flag:
    print("ì´ìƒì¹˜ê°€ ë°œê²¬ë˜ì–´ ìŠ¤í”¼ì–´ë§Œ ìƒê´€ê³„ìˆ˜ ì‚¬ìš©")
else:
    print("ì´ìƒì¹˜ê°€ ë°œê²¬ë˜ì§€ ì•Šì•„ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ ì‚¬ìš©")
```

```python
from pandas import DataFrame
from statsmodels.stats.outliers_influence import variance_inflation_factor

def ols_report(tbl, data):
    """
    OLS summary(tbl)ì™€ ë°ì´í„°í”„ë ˆì„(data)ì„ ë°›ì•„ íšŒê·€ê³„ìˆ˜ í‘œ(rdf),
    ëª¨ë¸ ì í•©ë„ ìš”ì•½(result_report), ëª¨ë¸ ë³´ê³  ë¬¸ì¥(model_report),
    ë³€ìˆ˜ë³„ ë³´ê³  ë¦¬ìŠ¤íŠ¸(variable_reports)ë¥¼ ë°˜í™˜í•œë‹¤.

    Parameters
    ----------
    tbl : statsmodels.iolib.summary.Summary
        sm.OLS(...).fit().summary() ê²°ê³¼ë¡œ ìƒì„±ëœ summary ê°ì²´.
    data : pandas.DataFrame
        ì¢…ì†ë³€ìˆ˜ì™€ ë…ë¦½ë³€ìˆ˜ë¥¼ ëª¨ë‘ í¬í•¨í•œ ë°ì´í„°í”„ë ˆì„.

    Returns
    -------
    tuple
        (íšŒê·€ê³„ìˆ˜ DataFrame, ì í•©ë„ ìš”ì•½ ë¬¸ìì—´, ëª¨ë¸ ë³´ê³  ë¬¸ìì—´, ë³€ìˆ˜ë³„ ë³´ê³  ë¦¬ìŠ¤íŠ¸)
    """

    # ë…ë¦½ë³€ìˆ˜ ì´ë¦„(ìƒìˆ˜í•­ ì œì™¸)
    xnames = [n for n in fit.model.exog_names if n != "const"]

    # ë…ë¦½ë³€ìˆ˜ ë¶€ë¶„ ë°ì´í„° (VIF ê³„ì‚°ìš©)
    indi_df = data.filter(xnames)

    # ë…ë¦½ë³€ìˆ˜ ê²°ê³¼ë¥¼ ëˆ„ì 
    variables = []
    for i, v in enumerate(tbl.tables[1].data):
        # í•œ í–‰ì˜ ë³€ìˆ˜ëª… ì¶”ì¶œ í›„ ëª©ë¡ì— ìˆëŠ”ì§€ í™•ì¸
        name = v[0].strip()
        if name not in xnames:
            continue

        # VIF ê³„ì‚°ì„ ìœ„í•´ ì—´ ì¸ë±ìŠ¤ ì¡°íšŒ
        j = list(indi_df.columns).index(name)
        vif = variance_inflation_factor(indi_df, j)

        # ìœ ì˜í™•ë¥ ê³¼ ë³„í‘œ í‘œì‹œ í•¨ìˆ˜
        p = float(v[4].strip())
        stars = lambda p: (
            "***" if p < 0.001 else
            "**"  if p < 0.01 else
            "*"   if p < 0.05 else
            ""
        )

        # í•œ ë³€ìˆ˜ì— ëŒ€í•œ ë³´ê³  ì •ë³´ ì¶”ê°€
        variables.append({
            "ì¢…ì†ë³€ìˆ˜": yname,                           # ì¢…ì†ë³€ìˆ˜ ì´ë¦„
            "ë…ë¦½ë³€ìˆ˜": name,                            # ë…ë¦½ë³€ìˆ˜ ì´ë¦„
            "B": v[1].strip(),                           # ë¹„í‘œì¤€í™” íšŒê·€ê³„ìˆ˜(B)
            "í‘œì¤€ì˜¤ì°¨": v[2].strip(),                    # ê³„ìˆ˜ í‘œì¤€ì˜¤ì°¨
            "Beta": float(fit.params[name])              # í‘œì¤€í™” íšŒê·€ê³„ìˆ˜(Î²)
                    * (data[name].std(ddof=1)
                       / data[yname].std(ddof=1)),
            "t": "%s%s" % (v[3].strip(), stars(p)),      # t-í†µê³„ëŸ‰(+ë³„í‘œ)
            "p-value": p,                                # ê³„ìˆ˜ ìœ ì˜í™•ë¥ 
            "ê³µì°¨": 1 / vif,                             # ê³µì°¨(Tolerance = 1/VIF)
            "vif": vif,                                  # ë¶„ì‚°íŒ½ì°½ê³„ìˆ˜
        })

    rdf = DataFrame(variables)

    # summary í‘œì—ì„œ ì í•©ë„ ì •ë³´ë¥¼ key-valueë¡œ ì¶”ì¶œ
    result_dict = {}
    for i in [0, 2]:
        for item in tbl.tables[i].data:
            n = len(item)
            for i in range(0, n, 2):
                key   = item[i].strip()[:-1]
                value = item[i+1].strip()
                if not key or not value:
                    continue
                result_dict[key] = value

    # ì í•©ë„ ë³´ê³  ë¬¸ìì—´ êµ¬ì„±
    result_report = (
        f"R({result_dict['R-squared']}), "
        f"R^2({result_dict['Adj. R-squared']}), "
        f"F({result_dict['F-statistic']}), "
        f"ìœ ì˜í™•ë¥ ({result_dict['Prob (F-statistic)']}), "
        f"Durbin-Watson({result_dict['Durbin-Watson']})"
    )

    # ëª¨í˜• ë³´ê³  ë¬¸ì¥ êµ¬ì„±
    tpl = (
        "%sì— ëŒ€í•˜ì—¬ %së¡œ ì˜ˆì¸¡í•˜ëŠ” íšŒê·€ë¶„ì„ì„ ì‹¤ì‹œí•œ ê²°ê³¼, "
        "ì´ íšŒê·€ëª¨í˜•ì€ í†µê³„ì ìœ¼ë¡œ %s"
        "(F(%s,%s) = %s, p %s 0.05)."
    )
    model_report = tpl % (
        rdf['ì¢…ì†ë³€ìˆ˜'][0],
        ",".join(list(rdf['ë…ë¦½ë³€ìˆ˜'])),
        "ìœ ì˜í•˜ë‹¤"
        if float(result_dict['Prob (F-statistic)']) <= 0.05
        else "ìœ ì˜í•˜ì§€ ì•Šë‹¤",
        result_dict['Df Model'],
        result_dict['Df Residuals'],
        result_dict['F-statistic'],
        "<" if float(result_dict['Prob (F-statistic)']) <= 0.05 else ">"
    )

    # ë³€ìˆ˜ë³„ ë³´ê³  ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸ êµ¬ì„±
    variable_reports = []
    s = "%sì˜ íšŒê·€ê³„ìˆ˜ëŠ” %s(p %s 0.05)ë¡œ, %sì— ëŒ€í•˜ì—¬ %s ì˜ˆì¸¡ë³€ì¸ì¸ ê²ƒìœ¼ë¡œ ë‚˜íƒ€ë‚¬ë‹¤."
    for i in rdf.index:
        row = rdf.iloc[i]
        variable_reports.append(
            s % (
                row['ë…ë¦½ë³€ìˆ˜'],
                row['B'],
                "<" if float(row['p-value']) < 0.05 else ">",
                row['ì¢…ì†ë³€ìˆ˜'],
                'ìœ ì˜ë¯¸í•œ'
                if float(row['p-value']) < 0.05
                else 'ìœ ì˜í•˜ì§€ ì•Šì€'
            )
        )

    return rdf, result_report, model_report, variable_reports
```

```python
mse_sq = fit.mse_resid
mse_root = np.sqrt(mse_sq)
r = fit.resid          # ì”ì°¨

r1 = r[(r > -mse_sq) & (r < mse_sq)].count() / r.count() * 100
r2 = r[(r > -2*mse_sq) & (r < 2*mse_sq)].count() / r.count() * 100
r3 = r[(r > -3*mse_sq) & (r < 3*mse_sq)].count() / r.count() * 100

mse_r = [r1, r2, r3]
colors = ["#0B3C5D", "#328CC1", "#D9EAF7"]

xmax = fit.fittedvalues.max()
ymin = fit.fittedvalues.min()

# 1) ê·¸ë˜í”„ ì´ˆê¸°í™”
width_px  = 1600
height_px = 720
figsize = (width_px / my_dpi, height_px / my_dpi)

fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=my_dpi, constrained_layout=True)

# 2) seaborn ê¸°ë³¸ ì”ì°¨ë„
sb.residplot(x=fit.fittedvalues, y=fit.resid)

# 3) ê¾¸ë¯¸ê¸°
ax.grid(True, alpha=0.3)

# 4) MSE 3ë‹¨ê³„ í‘œì‹œ
for i, c in enumerate(colors):
    k = i + 1

    # ìœ—ë°©í–¥ Â±kâˆšMSE ìœ„ì¹˜
    y_pos = k * mse_root

    ax.axhline(y=y_pos,  color=c, linestyle="--", alpha=0.6)
    ax.axhline(y=-y_pos, color=c, linestyle="--", alpha=0.6)

    # í…ìŠ¤íŠ¸ëŠ” ì¶• ì¢Œí‘œê³„ ì‚¬ìš© (ê·¸ë˜í”„ ë°–ìœ¼ë¡œ ì•ˆ ë‚˜ê°)
    ax.text(x=1.02, y=0.5 + 0.12*k,
            s=f"+{k} âˆšMSE = {mse_r[i]:0.2f}%",
            transform=ax.transAxes,
            ha="left", va="center", fontsize=11, color=c)

    ax.text(x=1.02, y=0.5 - 0.12*k,
            s=f"-{k} âˆšMSE = {mse_r[i]:0.2f}%",
            transform=ax.transAxes,
            ha="left", va="center", fontsize=11, color=c)

# 5) ì¶œë ¥
plt.show()
plt.close()
```

```python
if len(origin) < 5000:
    method = "Shapiro-Wilk"
    stat, p = shapiro(fit.resid)
    print(f"{method} statistic: {stat:.4f}, p-value: {p:.4f}")
    violated = p < 0.05
else:
    method = "Anderson-Darling"
    ad_res = anderson(fit.resid, dist='norm')
    crit_levels = ad_res.significance_level.tolist()
    crit_values = ad_res.critical_values.tolist()
    crit_5 = crit_values[crit_levels.index(5.0)] if 5.0 in crit_levels else None
    print(f"{method} statistic: {ad_res.statistic:.4f}, 5% ì„ê³„ê°’: {crit_5:.4f}")
    violated = crit_5 is not None and ad_res.statistic > crit_5

print("ì •ê·œì„± ê°€ì • ìœ„ë°° ê°€ëŠ¥ì„±ì´ ìˆìŒ (5% ìœ ì˜ìˆ˜ì¤€)" if violated else
      "ì •ê·œì„± ê°€ì • ìœ„ë°° ê·¼ê±° ì—†ìŒ (5% ìœ ì˜ìˆ˜ì¤€)")
```
