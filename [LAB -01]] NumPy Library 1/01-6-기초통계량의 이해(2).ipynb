### 라이브러리 참조하기

```python
import numpy as np
```

```python
point= np.array([100,91,89,86,84,79,78,77,74,71,69,66,65,60,58,57,55])

point
```

### Numpy 의 Histogram : 도수 분포표

```python
#histogram 을 사용하면 값이 2개가 반환되는데 각 구간별 데이터 수와 그 데이터들의 경계값 (경계값 데이터가 1개 더 많음)
hist,bins = np.histogram(point,5)
print("구간별 데이터 수", hist)
print("구간 경계:",bins)
```

```python
#도수 분포표의 구성

s = len(hist)
for i in range(0, s):
    if i + 1 < s:   #마지막 항목을 생략할 수 있음
        tpl = "%d이상 %d미만: %d개"
    else:
        tpl = "%d이상 %d이하: %d개"
    print(tpl % (bins[i], bins[i+1], hist[i]))
```

```python
#데이터 구간 직접 정의하기


hist, bins = np.histogram(point, [50, 60, 70, 80, 90, 100])
print("구간별 데이터 수:", hist)
print("구간 경계:", bins)
```

### 도수의 종류

```python
#상대 도수 (절대도수를 백분률로 환산한 값으로, 각 도수를 도수의 총 합으로 나눈 값)

relative_freq = hist / np.sum(hist)
relative_freq
```

```python
#누적 도수 (절대도수에서 자기 앞 도수를 모두 다 더한 값)


cum_freq = []

previous = 0

for i in hist:
    previous += i
    cum_freq.append(previous)

previous
```
