## 05 기초 통계량의 이해(1)
<기초 통계량의 종류>
- 도수 (데이터의 수)
- 최대값 (데이터 중 가장 큰 값)
- 최소값 (데이터 중 가장 작은 값)
- 합계 (모든 값을 더한 값)
- 평균 (전체의 대표값 / 데이터의 중심)
- 중앙값 (가운데 정렬 값) *극단값의 영향 적음
- 사분위 수 (데이터를 4등분하는 기준값) * 분포 형태 판단




### 라이브러리 참조하기

```python
import numpy as np

#샘플 데이터 생성
myarr = np.array([98,72,80,64,88,65,79,59,81,74])
myarr
```

### 데이터의 위치와 분포를 나타내는 기초 통계량

```python
#총 도수
print("총 도수:",len(myarr))
```

```python
# 최대,최소,합계
print("최대값", np.max(myarr))
print("최소값", np.min(myarr))
print("합계", np.sum(myarr))
```

```python
#평균
print("평균",np.average(myarr))
```

```python
#중앙값
print("중앙값",np.median(myarr))
```

```python
# 중앙값 직접 구하기

size = len(myarr)
sorted_arr=np.sort(myarr)
print("정렬",sorted_arr)

# 중앙값 계산
if size % 2 == 0:  # 짝수인 경우
    p = (size // 2) - 1    # 인덱스가 0부터 시작하므로 -1
    q = p + 1              # p번째 다음 요소
    median = (sorted_arr[p] + sorted_arr[q]) / 2
else:  # 홀수인 경우
    p = size // 2
    median = sorted_arr[p]

print("중앙값:", median)
```

```python
# 사분위 수

# 1사분위 수를 구하기 위해 우선 데이터 정렬
x = np.sort(myarr)      # 정렬된 배열
n = len(x)              # 총 도수(데이터 개수)

# 위치 계산
q1_pos = (n - 1) * 0.25   # Q1 위치
i = int(np.floor(q1_pos)) # 아래쪽 인덱스
f = q1_pos - i            # 보간 비율 (소수부분)

# 보간(interpolation)
Q1 = x[i] + f * (x[i + 1] - x[i])
print("직접 계산한 1사분위수 (Q1):", Q1)

# numpy 내장 함수로 비교
print("NumPy로 계산한 1사분위수:", np.quantile(myarr, 0.25))
```

```python
# 사분위수 구간 (IQR)
print("사분위수 구간(IQR):", np.quantile(myarr,0.75)-np.quantile(myarr,0.25))
```

```python
#분산 표준편차
print("분산",np.var(myarr))
print("표준편차",np.std(myarr))
```

### 2차원 배열의 기초 통계량
- axis 에 따라 열끼리 계산하는지 행끼리 계산하는지가 결정된다

```python
# 2차원 배열 생성
mydata = np.array([
    [98, 72, 80, 64, 88],
    [65, 79, 59, 81, 74]
])

# (1) 각 열끼리 연산 (세로 방향 연산)
s1 = np.sum(mydata, axis=0)
print(type(s1))
print(s1)   # 각 열의 합

# (2) 각 행끼리 연산 (가로 방향 연산)
s2 = np.sum(mydata, axis=1)
print(type(s2))
print(s2)   # 각 행의 합
```
