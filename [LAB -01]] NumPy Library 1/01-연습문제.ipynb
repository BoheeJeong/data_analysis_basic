```python

# 1. dtype íŒŒë¼ë¯¸í„°ë¡œ ë°ì´í„° íƒ€ì…ì„ ê°•ì œë¡œ ì§€ì •í•´ì¤„ ìˆ˜ ìˆê³ , int ë¡œ ì§€ì •í•˜ë©´ ì†Œìˆ˜ì ì€ ëª¨ë‘ ì‚¬ë¼ì§„ë‹¤

score = np.array ([72.5, 88.0, 91.5, 67.0, 80.0, 59.5],dtype=int)

print("ì›ë³¸ ì ìˆ˜ëŠ”",score)   

# 2. ì ìˆ˜ê°€ 80ì  ì´ìƒì¸ ê²½ìš°ë§Œ ì¶”ì¶œí•˜ê¸° ìœ„í•´ ndarray ë’¤ì— ì¡°ê±´ì„ ê±¸ì–´ì¤€ë‹¤
more80score = score[score>=80]
print("80ì  ì´ìƒ ì ìˆ˜ëŠ”",more80score)


# 3. ëª¨ë“  í•™ìƒì—ê²Œ íŠ¹ì • ê¸°ì¤€ìœ¼ë¡œ í•™ì ì„ ë¶€ì—¬í•˜ê³  ê²°ê³¼ë¥¼ ì¶”ì¶œí•˜ì‹œì˜¤

score_conditions =[ (score >=90),
                    (score >=80),
                    (score >=70)
]


score_value = ['A','B','C']

result=np.select(score_conditions,score_value,default='F')
print("ê° ì ìˆ˜ì— ëŒ€í•œ ë“±ê¸‰ì€" , result)
```

 3. 2ë²ˆì˜ ë„ìˆ˜ ë¶„í¬í‘œë¥¼ í™œìš©í•˜ì—¬ ì ˆëŒ€ë„ìˆ˜, ìƒëŒ€ë„ìˆ˜, ëˆ„ì ë„ìˆ˜ë¥¼ í¬í•¨í•˜ëŠ” ë„ìˆ˜ë¶„í¬í‘œë¥¼

```python
import numpy as np

daily_sales_data = np.array([18,25,33,21,28,40,15,3,9,17,22,29,38,45,11,6,31,24,27,35,12,19,39,42,7,2,20,34,26,30])



hist,bins = np.histogram(daily_sales_data,[0,10,20,30,40,50])
print("êµ¬ê°„ë³„ ë°ì´í„° ìˆ˜:" , hist) #ì ˆëŒ€ë„ìˆ˜
print("êµ¬ê°„ ê²½ê³„",bins)
```

```python
#ìƒëŒ€ë„ìˆ˜
relative_freq = hist/ np.sum(hist)
np.round(relative_freq,3)
```

```python
#ëˆ„ì  ë„ìˆ˜
cum_freq = []
previous = 0

for i in hist:
  previous +=i
  cum_freq.append(previous)
previous
```
