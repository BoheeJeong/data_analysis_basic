# 📘 01. 추론통계

## [1] 정의
모집단으로부터 추출된 표본의 통계량을 이용하여 **모집단의 특성(모수)** 을 추론하는 절차.

- 실질적인 **데이터 분석(통계)** 의 핵심 과정
- 표본 정보를 이용해 모집단에 대한 추측·결론을 이끌어내는 과정
- 수집된 자료를 이용해 모집단 전체에 대한 의사결정을 하는 것
- 제한된 표본을 기반으로 모집단에 대한 일반화된 결론을 유도하기 때문에 **본질적으로 불확실성**을 수반함

---

## [2] 추론 통계의 결론

- 성별에 따른 월급 차이가 **우연히 나타날 확률이 작다**  
  → **통계적으로 유의하다(statistically significant)**

- 성별에 따른 월급 차이가 **우연히 나타날 확률이 크다**  
  → **통계적으로 유의하지 않다(not statistically significant)**

즉, 추론 통계 분석을 수행했다는 것은 **가설 검정을 수행했다**는 의미이다.

# 📘 02. 데이터마이닝(머신러닝)

## 1) 개요
- 대표적인 고급 데이터 분석법
- 대용량 자료로부터 유용한 정보를 요약
- 미래에 대한 예측
- 관계, 패턴, 규칙 탐색
- 모형화
- 유용한 지식을 추출

---

## 2) 방법론

### 🔹 데이터베이스에서의 지식탐색 (KDD)
- 데이터웨어하우스에서 데이터마트를 생성
- 각 데이터들의 속성을 사전 분석하여 지식을 얻음

### 🔹 기계학습(Machine Learning)
- 컴퓨터가 학습할 수 있도록 알고리즘과 기술을 개발하는 분야
- 예시:
  - 인공지능, 의사결정나무, 클러스터링, 베이지안 분류, SVM 등

---

## 3) 활용 분야

| 분야 | 예시 |
|------|------|
| 데이터베이스 마케팅 | 방대한 고객 행동정보 활용 (예: 목표 마케팅, 고객세분화, 장바구니 분석 등) |
| 신용평가 / 조기경보 시스템 | 신용카드 발급, 보험, 대출 업무 등 |
| 생명정보학 | 유전자 분석, 질병 진단, 치료법/신약 개발 |
| 텍스트마이닝 | 전자우편, SNS, 디지털 텍스트 기반 고객 분석·감성 분석·사회관계망 분석 |
