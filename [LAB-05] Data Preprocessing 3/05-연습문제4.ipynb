## 04 명목형 변수 인코딩

<연습문제>
□ 한 온라인 스토어의 마케팅 팀이 연말 특별 프로모션을 기획하고 있습니다.

프로모션의 효과를 극대화하기 위해, 고객 데이터를 분석하여 타겟 고객층을 식별하고자 합니다.

여러분은 데이터 분석가로서 마케팅 팀의 의사결정을 돕기 위해 고객 데이터를 전처리하고 분석용 데이터셋을 만드는 임무를 맡았습니다.

□ 주어지는 데이터 셋

online_store_customers : 고객의 기본 인구 통계 정보 (고객ID, 성별, 나이, 거주 도시, 멤버십 등급)

online_store_purchases : 고객의 구매 관련 정보 (고객ID, 마지막 구매일, 총 소비 금액, 이탈 여부)


해야 할 작업

1. 두 데이터를 불러와서 고객 ID를 기준으로 병합하세요.

2. 구매 데이터의 size 열에 존재하는 결측치는, 가장 빈도가 높은 사이즈로 채워주세요.

3. 성별에 따른 구매 패턴을 분석하기 위해 gender 열에 대해 더미 변수를 생성하세요.

4. 색상 선호도를 분석하기 위해 color 열을 원-핫 인코딩(One-hot encoding) 방식으로 변환하여 새로운 열들을 추가하세요.

```python
from hossam import load_data
from pandas import DataFrame, get_dummies, merge
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import (
    MinMaxScaler, MaxAbsScaler, StandardScaler,
    RobustScaler, Normalizer
)
```

```python
online_store_customers =load_data('online_store_customers')
online_store_customers

```

```python
online_store_purchases = load_data('online_store_purchases')
online_store_purchases
```

```python
# step 1 : index 와 컬럼을 기준으로 병합한다
merged_df = merge (online_store_customers,online_store_purchases,left_index=True,right_on="user_id")
merged_df
```

```python
# step 2 : size 의 결측치는 가장 빈도가 높은 사이즈로 채워주세요

most_frequent_size = merged_df['size'].mode()

merged_df_copy=merged_df.copy()

#most_frequent_size 가 SERIES 이기 떄문에 인덱스 번호를 붙여줘야한다
merged_df_copy['size'] = merged_df_copy['size'].fillna(value=most_frequent_size[0])
merged_df_copy

```

```python
# step 3. 성별에 따른 구매 패턴을 분석하기 위해 gender 열에 대해 더비 변수를 생성하세요

merged_df_copy2 = merged_df_copy.copy()

merged_df_copy3=get_dummies(merged_df_copy2, columns=['gender'],dtype='int',drop_first=True)

merged_df_copy3
```

```python
# step 4. 색상 선호도를 분석하기 위해 color 열을 원핫 인코딩 방식으로 변환하여 새로운 열들을 추가하세요

# 4-1 인코딩 처리할 명목형 변수의 차원 변환을 수행하여 데이터 프레임을 유지한다

X=merged_df_copy3['color'].values.reshape(-1,1)
X
```

```python
# 4-2 원핫 인코딩 처리
encoder = OneHotEncoder(sparse_output=False , drop = None)
result=encoder.fit_transform(X)
result
```

```python
# 4-3 인코딩 결과를 데이터 프레임으로 생성
#컬럼명 생성

new_cols = encoder.get_feature_names_out(['color'])
new_cols
```

```python
#DataFrame 병합
one_hot_df = DataFrame(result,columns=new_cols,index=merged_df_copy3.index)
one_hot_df.head()
```

```python
merged_df_copy4=merged_df_copy3.copy()


result_df = merge(merged_df_copy4,one_hot_df,left_index=True,right_index=True,how='left')
result_df
```
