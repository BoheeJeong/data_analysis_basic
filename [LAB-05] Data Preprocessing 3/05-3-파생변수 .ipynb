### #01.준비작업

```python
#라이브러리 참조
from hossam import load_data
import numpy as np
```

```python
#데이터 가져오기
origin = load_data('grade')
origin
```

### #02.파생변수 추가하기
- 컬럼 연산을 통한 새로운 컬럼 추가

```python
df1 = origin.copy()

#직접 더해주는 방식은 나중에 갯수가 늘어나면 힘들어질 수 있음
df1['직접합계']=df1['국어']+df1['영어']+df1['수학']+df1['과학']

df1
```

### #02.파생변수 추가하기 
- 기술통계 관련 함수를 사용한 파생변수
- 컬럼 이름에 대한 슬라이싱 접근

```python
df2 = origin.copy()

#기존에 없던 컬럼 이름을 넣으면 바로 새로운 컬럼을 만들 수 있다
#axis =1 이면 x 축 0 이면 y 축 계산을 의미한다
df2['합계']=df2[df2.columns[2:6]].sum(axis=1)
df2
```

### #02.파생변수 추가하기 
- 기술통계 관련 함수를 사용한 파생변수
- 인덱스 기반 위치로 슬라이싱 하여 합계 변수 추가

```python
df3 = origin.copy()

#모든 행에 대해서 , 2번째 컬럼부터 끝까지, x 축으로 합계 산출
df3['합계']=df3.iloc[:,2:].sum(axis=1)
df3
```

### #02.파생변수 추가하기 
- 기술통계 관련 함수를 사용한 파생변수
- 나머지 기술 통계값 추가

```python
# 평균값 추가
df3['평균'] = df3.iloc[:, 2:6].mean(axis=1)

# 최대값 추가
df3['최대값'] = df3.iloc[:, 2:6].max(axis=1)

# 최소값 추가
df3['최소값'] = df3.iloc[:, 2:6].min(axis=1)

# 표준편차 추가
df3['표준편차'] = df3.iloc[:, 2:6].std(axis=1)

# 분산 추가
df3['분산'] = df3.iloc[:, 2:6].var(axis=1)

df3
```

### #03.조건별 파생 변수
- if else

```python
#실습용 합계와 평균 파생변수 추가
df4=origin.copy()
df4['평균'] = df4.iloc[:, 2:6].mean(axis=1)
df4['합계'] = df4.iloc[:, 2:6].sum(axis=1)
df4
```

### #03.조건별 파생 변수
- if else : Numpy의 where 함수를 사용한 조건부 파생 변수

```python
#80점이 넘으면 합격, 아니면 불합격 
df4['결과1'] = np.where(df4['평균']>=80,'합격','불합격')
df4
```

### #03.조건별 파생 변수
- if else : 표현식을 사용한 평균값에 따른 조건부 파생 변수

```python
#디폴트 값 불합격을 설정해두고 80점이 넘은 경우에만 합격으로 입력되도록
df4['결과2']='불합격'
df4.loc[df4['평균']>=80,'결과2']='합격'
df4
```

### #03.조건별 파생 변수
- elif : NumPy 의 select 함수 사용

```python
# 학점을 부여하기 위한 점수의 구간을 설정하는 조건들을 리스트로 설정
conditions = [
    (df4['평균'] >= 90),   # if
    (df4['평균'] >= 80),   # elif
    (df4['평균'] >= 70)    # elif
]

# 조건에 따라 부여될 학점
value = ['A', 'B', 'C']

#conditions 조건을 확인한 다음, 그 조건에 따라 각각 value 값을 할당하고, 기본 값은 F 로 설정
df4['학점1'] = np.select(conditions, value, default='F')

df4
```

### #03.조건별 파생 변수
- 사용자 정의 함수 사용 : 함수 정의하기
- 정의한 함수 리턴값으로 파생변수 추가하기

```python
def getScore(x):
  if x>= 90:
    score ='A'
  elif x>= 80:
    score = 'B'
  elif x>= 70:
    score = 'C'
  else :
    score = 'F'


  return score
```

```python

#학점 2 라는 컬럼을 새로 만들 것인데, df4 에 있는 평균 컬럼에 getScore 함수를 적용한 값들을 출력합니다
df4['학점2']=df4['평균'].apply(getScore)
df4
```

### #03.조건별 파생 변수
- 함수 정의 없이 람다식으로 익명 함수 적용하기

```python
df4['학점3'] = df4['평균'].apply(lambda x: {10:'A',9:'A',8:'B',7:'C'}.get(x//10,'F'))
df4
```

### #03.조건별 파생 변수
-map() 함수 사용 : 값이 완전 일치하는 경우에만 활용 가능

```python
df4['성별코드']=df4['성별'].map({'남자':'M','여자':'F'})
df4
```
