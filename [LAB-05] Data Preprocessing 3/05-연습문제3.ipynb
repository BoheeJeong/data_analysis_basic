```python
#Ìå®ÌÇ§ÏßÄ Ï∞∏Ï°∞
from hossam import load_data
from pandas import DataFrame
```

```python
online_shop_sales = load_data('online_shop_sales')
online_shop_sales
```

```python
#Í∞Å Í≥†Í∞ù(UserID)Î≥Ñ Ï¥ù Íµ¨Îß§ Í∏àÏï°(Total) ÏùÑ Í≥ÑÏÇ∞ÌïòÏÑ∏Ïöî.

# Ï¥ù Íµ¨Îß§ Í∏àÏï°ÏùÄ Í∞úÎ≥Ñ Ï£ºÎ¨∏Ïùò Price √ó Quantity Í∞íÏùÑ Î™®Îëê ÎçîÌïú Í∏àÏï°ÏûÖÎãàÎã§.
online_shop_sales_copy = online_shop_sales.copy()

online_shop_sales_copy['Total']=online_shop_sales_copy['Price']*online_shop_sales_copy['Quantity']

online_shop_sales_copy
```

```python
online_shop_sales_copy2=online_shop_sales_copy.filter(['UserID','Total'])
online_shop_sales_copy2
```

```python
#Í∞Å Í≥†Í∞ù ID Î≥Ñ Ï¥ù Íµ¨Îß§ Í∏àÏï°
online_shop_sales_copy3=online_shop_sales_copy2.groupby(['UserID']).sum()
online_shop_sales_copy3
```

```python
# soflacktns wjdfuf
sorted= online_shop_sales_copy3.sort_values('Total',ascending=False)

sorted
```
