도시와 농촌별에 따른 남녀 평균 흡연율을 확인할 수 있는 데이터 피벗 테이블로 구성하시오.

```python
pivot_df=pivot_table(students_smoke, index = '지역',columns='구성',values='흡연율',aggfunc='mean')

pivot_df
```

<연습 문제2>
당신은 최근 동네에 새로 생긴 ‘달콤스쿱’ 아이스크림 가게의 데이터 분석을 의뢰받았습니다.

◎ 사장님은 여름 시즌을 맞아 어떤 맛 아이스크림이 인기가 많은지,
날짜별 판매 추이는 어떤지 궁금해하며,
효율적인 재고 관리를 위한 데이터 분석을 당신에게 요청했습니다.

◎ 당신에게 주어진 icecream_sales 데이터셋을 사용하여 다음 과제를 완수하세요.



1. 데이터의 전체적인 구조와 정보를 확인하고, 판매 수량에 결측치가 있는지 확인하세요.

만약 결측치가 있다면, 해당 행을 제거하세요.

2. 판매 날짜를 인덱스로, 아이스크림 맛을 컬럼으로 하여
각 맛의 일별 판매량을 보여주는 피벗 테이블을 만드세요.

3. 사장님이 다른 분석 툴에서 사용하고 싶다며,

위에서 만든 피벗 테이블을 다시 긴 형식(Long format) 의 데이터로 변환해 달라고 요청했습니다.
이때 컬럼 이름은 원본 이름을 그대로 유지해야 합니다.

```python
icecream_sales = load_data('icecream_sales')
icecream_sales 
```

```python
#데이터의 전체적인 구조와 정보를 확인하고, 판매 수량에 결측치가 있는지 확인하세요. 결측치는 제거하세요
icecream_sales_copy1=icecream_sales.copy()
icecream_sales_copy1.info()
```

```python
rows,cols = icecream_sales_copy1.shape
print(f'{rows} 행, {cols} 열,{icecream_sales_copy1.ndim} 차원의 데이터 ' )
```

```python
#nan 행 삭제
icecream_sales_copy2_행삭제 = icecream_sales_copy1.copy()

icecream_sales_copy2_행삭제결과 = icecream_sales_copy2_행삭제.dropna(subset=['Quantity'])
icecream_sales_copy2_행삭제결과
```

```python
# #2. 판매 날짜를 인덱스로, 아이스크림 맛을 컬럼으로 하여
# 각 맛의 일별 판매량을 보여주는 피벗 테이블을 만드세요.


icecream_pivot = pivot_table(icecream_sales_copy2_행삭제결과,index='Date',columns='Flavor',values='Quantity')
icecream_pivot
```

```python

# 3. 사장님이 다른 분석 툴에서 사용하고 싶다며,

# 위에서 만든 피벗 테이블을 다시 긴 형식(Long format) 의 데이터로 변환해 달라고 요청했습니다.
# 이때 컬럼 이름은 원본 이름을 그대로 유지해야 합니다.

#1단계 > 데이터 프레임으 인덱스를 일반 컬럼으로 설정한다

icecream_pivot_copy=icecream_pivot.copy()

icecream_pivot_copy_2 = icecream_pivot_copy.reset_index()
icecream_pivot_copy_2

```

```python
mdf = melt(icecream_pivot_copy_2,id_vars=['Date'],value_vars=['딸기','민트초코','바닐라','초콜릿'],var_name='Flavor',value_name='Quantity')

mdf
```
