## 01 데이터 재구조화 - Stack / Unstack

- Stack 은 행 인덱스 방향의 데이터를 열에서 행으로 압축하여 멀티인덱스 구조로 묶는 변환
- Unstack 은 멀티 인덱스로 구성된 행 인덱스를 행에서 열로 펼쳐 새로운 컬럼 구조를 만드는 변환 


### #.01 준비작업


```python
#라이브러리 참조
from hossam import load_data
from pandas import DataFrame
```

```python
origin = load_data('body_size')
origin
```

### #02 Stack


```python
st1 = origin.stack() #인덱스 열에 따라 모든 변수를 하나의 변수로 쌓는 처리 > 리턴 결과는 시리즈 객체 
print(type(st1))  
st1
```

```python
#stack 결과를 DataFrame 으로 생성하기 > NaN 값은 자동으로 빠진다
df1 = DataFrame(origin.stack()) 
df1
```

```python
#빈 값 포함 시키기 > dropna= False 옵션으로 NaN 을 그대ㅗㄹ 유지할 수 있다

st2 = origin.stack(dropna=False)
df2=DataFrame(st2)
df2
```

### #03 Unstack > stack 결과를 원래대로 되돌린다


```python
st2.unstack()
```
