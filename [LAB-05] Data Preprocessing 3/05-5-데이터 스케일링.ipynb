# LAB 05 - Data preprocessing

## 05 데이터 스케일링


<데이터 스케일링>
- 연속형 변수 > 스케일링 
- 변수의 크기를 조정
- 변수 단위가 다른 경우 모델이 큰 값 위주로 학습
- 더미변수에는 스케일링 불필요

<스케일링 기법 비교>
- MinMaxScaler : 이상치 영향 큼
- MaxAbsScaler : 영향 중간
- StandardScaler : 영향 중간 (선형회귀,로지스틱 회귀)
- RobustScaler : 영향 적음


### #01.준비 작업

```python
#라이브러리 참조
from hossam import load_data
from sklearn.preprocessing import MinMaxScaler,MaxAbsScaler,StandardScaler,RobustScaler
```

```python
origin = load_data('nursing_grades')
origin
```

### #02.MinMax Scaler 
- 최소값을 0 , 최대값을 1로 변환 > 데이터 범위를 0-1 로 압축하여 사용 (분포 형태 보존)

### 직접 계산하기

```python

minmax_df = origin.copy()


x=minmax_df['필기점수']
xmin = x.min()
xmax=x.max()

minmax_df['필기점수_MinMax(1)']=(x-xmin) / (xmax-xmin)
minmax_df.head()
```

### Sklearn 활용

```python
#특정 필드만 추출
X= minmax_df[['필기점수']]   #2차원데이터 프레임 형태
X.head()
```

```python
#스케일링 적용
scaler = MinMaxScaler()
minmax_df['필기점수_MinMax(2)'] = scaler.fit_transform(X)    #MINMAXSCALER 의 객체를 만들어서 추출한 관측값 데이터를 fit_tranform 으로 던져주기


minmax_df.head()
```

### #03.Max-Abs Scaler
- 절대값 기준으로 -1 ~ 1 범위로 스케일링 
- 음수와 양수 모두 유지하고, 희소 행렬에 유리
- 희소특성 데이터에 추천
- 이상치 영향 중간

### 직접 계산하기

```python
abs_df = origin.copy()
x = abs_df['필기점수']
abs_df['필기점수_MaxAbs(1)'] = x/x.abs().max()
abs_df.head()
```

### Sklearn

```python
scaler = MaxAbsScaler()
x=abs_df[['필기점수']]
abs_df['필기점수_MaxAbs(2)'] = scaler.fit_transform(x)
abs_df.head()
```

### #04.StandardScaler (z-score , 표준화)
- 데이터를 평균이 0 , 표준편차가 1인 표준정규분포를 따르도록 변환
- 데이터를 정규분포에 근사시켜 이상치에 덜 민감하게 만들어준다
- 

### 직접 계산하기

```python
std_df = origin.copy()
x=origin['학부성적']
std_df['학부성적_Standard(1)'] = (x-x.mean())/x.std()
std_df.head()
```

### Sklearn

```python
scaler = StandardScaler()
x=std_df[['학부성적']]

std_df['학부성적_Standard(2)'] = scaler.fit_transform(x)
std_df.head()
```

### #04.RobustScaler
- 이상치가 존재할 경우 이상치의 영향을 최소화 하여 데이터를 스케일링 하는 방법
- 이상치가 포함된 데이터를 표준화 혹은 정규화를 할 떄 이상치의 영향으로 전체 데이터의 분포가 왜곡될 수 있음
- RobustScaler 는 이 문제를 해결하기 위해 중앙값과 사분위수를 사용하여 데이터를 스케일링 함

### 직접 계산하기

```python
rb_df = origin.copy()

x=rb_df['병원경력']
m=x.median()

iqr = x.quantile(0.75) - x.quantile(0.25)


rb_df['병원경력_Robust(1)'] = (x-m) / iqr
rb_df.head()
```

### Sklearn 활용

```python
scaler = RobustScaler()
x=rb_df[['병원경력']]
rb_df['병원경력_Robust(2)']=scaler.fit_transform(x)
rb_df.head()
```
