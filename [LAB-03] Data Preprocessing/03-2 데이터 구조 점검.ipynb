### #03.행/열 추가

<행 추가>
*데이터 프레임에서 행 접근은 a.loc[]
- 리스트로 추가하기 : 기존의 컬럼 순서대로 넣어야하고, 누락되는 값이 있거나 초과할 경우 에러
- 딕셔너리로 추가하기 :컬럼 순서 상관없이 넣을 수 있지만, 누락되는 값이 있거나 초과할 경우는 에러
- Series 객체로 추가하기 : 컬럼 순서 상관 없지만, Series 객체의 index 가 datagrame 의 변수명과 동일해야함
- 빈 값을 저장하는 경우 : 데이터를 추가할 떄, 값이 없는 필드는 생략할 수 있고, Series 형태로 객체를 추가할 때만 가능함
- 기존의 행을 복사해서 넣는 경우
- 단일 값 추가하기

```python
#행 추가하기 - 리스트로 추가하기
리스트_행추가df = origin.copy()
리스트_행추가df.loc['정호'] = [2,'남자',90,80,70,60] 
리스트_행추가df
```

```python
#행 추가하기 - 딕셔너리로 추가하기
# 원본 복사
딕셔너리_행추가df = origin.copy()

# 새 행 추가 (index = '민정') > 컬럼의 순서는 상관이 없는 것 확인
딕셔너리_행추가df.loc['민정'] = {
    '성별': '여자',
    '학년': 4,
    '국어': 81,
    '영어': 72,
    '수학': 84,
    '과학': 90
}

# 결과 출력
딕셔너리_행추가df
```

```python
#행 추가하기 - Series 객체로 추가하기

시리즈_행추가df = origin.copy()
s=Series([2,'남자',90,80,70,60],index=['학년','성별','국어','영어','수학','과학'])

시리즈_행추가df.loc['형석']=s
시리즈_행추가df
```

```python
#행 추가하기 - 빈 값을 저장해야 하는 경우
빈값_행추가df = origin.copy()
s=Series([2,'남자',93,71],index =['학년','성별','국어','수학'])   #데이터를 추가할 떄 값이 없는 필드를 생략하면 NaN 값으로 들어간다

빈값_행추가df.loc['석호']=s    
빈값_행추가df
```

```python
#행 추가하기 - 기존의 행을 복사해서 추가하는 경우

행_복사df = origin.copy()
행_복사df.loc['철민']=행_복사df.loc['철수']  #새로운 행을 생성해서 거기에 넣어준다
행_복사df
```

```python
#행 추가하기 - 단일값 추가하기
단일값_행추가df = origin.copy()
단일값_행추가df.loc['희영']=100   #모든 변수에 같은 값이 저장된다
단일값_행추가df
```

### #03.행/열 추가

<열 추가>
*데이터 프레임에서 열 접근은 a[]
- 리스트로 추가하기 : 모든 행에 대한 값을 순서대로 나열, 기존 행과 개수가 다르면 에러, 빈값을 추가하려 하면 None 명시
- 딕셔너리로 추가하기 : 추가하지 않을 행은 명시하지 않으면 NaN 으로 설정된다, 기존 dataframe 에 없는 인덱스면 추가되지 않는다
- Series 객체로 추가하기 : 각 값이 연결될 행의 이름 index 를 지정해야한다, 부분적으로 값을 비워둘 수 있다
- 기존의 행을 복사해서 넣는 경우 : 각 값이 연결될 행의 이름 index 를 지정해야한다
- 단일 값 추가하기 : 

```python
#열 추가하기 - 리스트로 추가하기
리스트_열추가df = origin.copy()
리스트_열추가df['한국사'] = [72,56,90,None,92,90,75,80,60] #빈값을 추가하고자 한 위치에 None 명시
리스트_열추가df
```

```python
#열 추가하기 - 딕셔너리로 추가하기
딕셔너리_열추가df = origin.copy()
딕셔너리_열추가df['세계사'] = {
    "철수": 90,
    "영희": 88,
    "민철": 72,
    "수현": 56,
    "호민": 65,    #원본 df 에는 없는 익덱스이므로 추가되지 않는다
    "용식": 77,
    "나영": 95,
    "석영": 54
}

# 결과 출력
딕셔너리_열추가df

#호영, 영호는 명시되지 않아 NaN 으로 빈값이 나온다
```

```python
#열 추가하기 - Series 객체로 추가하기
시리즈_열추가df = origin.copy()
s=Series([72,56,90,88] , index =['철수','민철','호영','용식'])
시리즈_열추가df['윤리'] = s
시리즈_열추가df
```

```python
#열 추가하기 - 기존의 열을 복사해서 추가하는 경우
열_복사df = origin.copy()
열_복사df['체육'] = 열_복사df['과학']
열_복사df
```

```python
# 열 추가하기 - 단일 값 추가하기
단일_값추가df = origin.copy()
단일_값추가df['체육'] = 100
단일_값추가df
```

### #04.행열 삭제하기

<행 삭제>
- 단일 행 삭제
- 다중 행 삭제
- index 에 대한 인덱싱으로 행 삭제
- index 에 대한 슬라이싱으로 행 삭제

```python
# 1) 행 삭제 - 단일 행 삭제
행_삭제df = origin.copy()
행_삭제df = 행_삭제df.drop(index = '철수')
행_삭제df
```

```python
# 2) 행 삭제 - 다중 행 삭제
다중_삭제df = origin.copy()
다중_삭제결과df = 다중_삭제df.drop(index = ['철수','민철','수현'])
다중_삭제결과df
```

```python
# 3) index 에 대한 익덱싱으로 행 삭제
인덱싱_행_삭제df = origin.copy()
인덱싱_행_삭제df.index
```

```python
인덱싱_행_삭제결과df = 인덱싱_행_삭제df.drop(index = 인덱싱_행_삭제df.index[0])
인덱싱_행_삭제결과df   # index[0] 자리가 사라짐
```

```python
# 4) index 에 대한 슬라이싱으로 행 삭제
다중_행_삭제df = origin.copy()
print(다중_행_삭제df.index)
```

```python
다중_행_삭제결과df = 다중_행_삭제df.drop(index = 다중_행_삭제df.index[0:7]) #인덱싱 0-6 자리 삭제
다중_행_삭제결과df
```

<열 삭제>
- 단일 열 삭제
- 다중 열 삭제
- 컬럼명에 대한 인덱싱으로 행 삭제
- 컬럼명 에 대한 슬라이싱으로 행 삭제

```python
# 1) 열 삭제 - 단일 열 삭제
단일열_삭제df = origin.copy()
단일열_삭제결과df = 단일열_삭제df.drop(columns = '학년')
단일열_삭제결과df
```

```python
# 2) 열 삭제 - 다중 열 삭제

다중_열_삭제df = origin.copy()
다중_열_삭제결과df = 다중_열_삭제df.drop(columns=['학년','성별','과학'])
다중_열_삭제결과df
```

```python
# 3) 열 삭제 - 컬럼 명에 대한 인덱싱으로 열 삭제

인덱싱_열_삭제df = origin.copy()
print(인덱싱_열_삭제df.columns)
```

```python
인덱싱_열_삭제결과df = 인덱싱_열_삭제df.drop(columns=인덱싱_열_삭제df.columns[0])
인덱싱_열_삭제결과df
```

```python
# 4) 열 삭제 - 컬럼 명에 대한 슬라이싱으로 열 삭제

다중_열_삭제df = origin.copy()
print(다중_열_삭제df.columns)
```

```python
다중_열_삭제결과df = 다중_열_삭제df.drop(columns = 다중_열_삭제df.columns[1:-2])
다중_열_삭제결과df
```
