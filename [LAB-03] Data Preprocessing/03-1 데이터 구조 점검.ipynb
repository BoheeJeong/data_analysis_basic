### #03.데이터 타입 변경
- 데이터를 핸들링 할 때는 항상 복사본에서 처리해준다
- 어떤 처리를 하고 새로 만들어진 데이터 프레임인지 한글로 명시해주면 이후 관리에 용이하다

```python
#컬럼 단위 변경을 위한 복사본 생성
컬럼타입변경df = origin.copy()
컬럼타입변경df.info()
```

```python
#개별 컬럼의 데이터 타입을 바꿀때는 시리즈 개체이므로 astype 함수로 변경 가능

컬럼타입변경df['국어']=컬럼타입변경df['국어'].astype('float')  
#변경된 결과는 다시 기존 컬럼에 저장해야 반영된다

컬럼타입변경df.info()
```

```python
#데이터 타입을 변경할 때는 딕셔너리를 활용해서 일괄적으로 수정도 가능하다

복수타입변환df = origin.astype({'국어':'float','영어':'float','수학':'float','과학':'float'})

복수타입변환df.info()
```

```python
#위에서 타입 변환을 하였어도 복사본에 반영된 것이기 떄문에 원본에는 변화가 없다
origin.info()
```

```python
#범주형 데이터를 명목 척도로 변환하기
#명목 척도로 변환해두어야 나중에 집계할 때 해당 변수는 연산되지 않는다


컬럼단위_명목변환df = origin.copy() #복사본 생성해주기


#개별 컬럼 단위로 명목척도 변환할 떄는 astype.category
컬럼단위_명목변환df['학년'] = 컬럼단위_명목변환df['학년'].astype('category')
컬럼단위_명목변환df['성별'] = 컬럼단위_명목변환df['성별'].astype('category')


컬럼단위_명목변환df.info()
```

```python
#복수개의 범주형 데이터를 명목척도로 변환할 때는 딕셔너리 활용 가능

#astype을 사용하면 복사본을 바로 리턴받을 수 있다
복수단위_명목변환df = origin.astype({'학년':'category','성별':'category'})

복수단위_명목변환df.info()
```

```python
#범주형 데이터를 서열 척도로 변환

#개별 컬럼 단위로 적용하여 원본에 직접 적용하기
컬럼단위_서열척도df = origin.copy()

#범주형 값의 종류를 리스트에 담고, 해당 값들이 순서가 있는 서열 척도임을 ordered 로 표현 가능
컬럼단위_서열척도df['학년']=컬럼단위_서열척도df['학년'].astype (CategoricalDtype(categories=[1,2,3,4],ordered=True))



컬럼단위_서열척도df.info()
```

```python
#범주형 데이터를 서열 척도로 변환 : DataFrame 전체에 적용후 복사본 리턴받기

서열척도df = origin.astype({
  '학년':CategoricalDtype(categories=[1,2,3,4],ordered=True)
})

서열척도df.info()
```

### #04.행/열의 이름 변경
- rename() 메서드를 이용
- before: after 형태로 딕셔너리를 작성
- 원본에는 변화 없음
- columns 와 indesc 중 원하는 파라미터 하나만 설정 가능
- 원본에 반영을 원하면 inplace=True


```python
이름변경df = origin.rename(
  columns ={'국어':'kor','영어':'eng','수학':'math','과학':'sci'},
  index ={'석영':'suk','호영':'ho','민철':'min','수현':'su'}
)

이름변경df
```
