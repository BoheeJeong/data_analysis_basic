#### 4-1 패키지 참조

```python
from IPython.display import Image
from wordcloud import WordCloud

#한글 형태소 분석기
from konlpy.tag import Okt

# 패키지의 데이터관련 정보를 갖는 객체 --> 설치 위치 파악용
from konlpy import data

# 파이썬 내장 클래스 --> 설치 불필요
from collections import Counter
import os
```

```python
nlp = Okt()
```

#### 4-3 샘플 문장에 대한 형태소 분석 수행

```python
sample = """개봉했을때부터 지금까지 마음이 답답하거나 힘들때
이영화 보고있어요 그때마다 심적인 위로를 받을수있는 영화같아요
장면 하나하나가 너무예쁘고 마음에 남아서 진한 여운까지 주는영화 감사합니다.
알잘딱깔센 위쑤시개 얼죽아"""


# 명사들만 추출 -> 리스트형식으로 반환
nouns = nlp.nouns(sample)
print(nouns)
```

### #03.대한민국 헌법 형태소 분석
#### 3-1 분석할 데이터 가져오기

```python
with open('C:\Data_Analysis_Basic\[LAB-08] 워드 클라우드\wordcloud\대한민국헌법.txt','r',encoding='utf-8') as f:
  text = f.read()

  print(text)
```

#### 3-2 데이터에 대한 형태소 분석

```python
# 형태소 분석 클래스의 객체 생성
nlp = Okt()

# 명사들만 추출 -> 리스트형식으로 반환
nouns = nlp.nouns(text)
 # print(nouns)
```

#### 3-3 불용어 제거하기

```python
#불용어 목록 가져오기
with open('wordcloud\stopwords-ko.txt','r',encoding = 'utf-8') as f:
  stopwords = f.readlines()

  for i ,v in enumerate(stopwords):
    stopwords[i] = v.strip()


  print(stopwords)
```

```python
#추출된 명사에서 불용어 제거

data_set = []


for v in nouns:
  if v not in stopwords:
    data_set.append(v)


print(data_set)
```

```python
#리스트로부터 빈도수 측정
count = Counter(data_set)
count
```

### 4.단어 빈도수 시각화

```python
#워드 클라우드 객체 생성 > 한글은 폰트 파일의 경로로 지정해야함

wc = WordCloud(
    font_path="./wordcloud/NanumMyeongjo.ttf",
    width=1200,
    height=800,
    scale=2.0,
    background_color="#ffffff",
    colormap="PuBu",

)

#직접 생성한 빈도수를 이용해서 워드 클라우드를 그린다

wc = wc.generate_from_frequencies(count)
wc.to_file('대한민국헌법-주요단어.png')
Image(filename = '대한민국헌법-주요단어.png')
```
