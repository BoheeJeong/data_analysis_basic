### 2.지도 표현하기

```python
#새로운 지도 객체 생성
map_osm2 = folium.Map(location = MY_PLACE , zoom_start= 17)


#마커 객체 생성
# -> 마커가 표시될 위치 [위도,경도] , 클릭 시 표시될 메세지, 아이콘 색상
my_marker = folium.Marker(MY_PLACE,popup='아이티윌 교육센터',icon = folium.Icon(color='darkred'))


my_marker.add_to(map_osm2)   #마커 객체를 지도에 추가함

#웹 페이지 파일이 저장될 폴더 생성
if not path.exists('output'):
  mkdir('output')

#지도가 표시된 웹 페이지 파일이 저장된다
#결과 확인은 윈도우 폴더 창에서 직접 웹 페이지 파일 더블 클릭
map_osm2.save('output/map_osm2.html')

```

### 3. (예제) 서울의 고등학교 분포 확인하기
#### 3-1 데이터 준비하기
전국 초중고 학교의 위치 데이터 로드

```python
origin = load_data('schools')
origin.head()
```

#### 3-2 데이터 전처리
(1) 사용할 필드만 추출

```python
df = origin.filter(['학교명','학교급구분','소재지도로명주소','위도','경도'])
df
```

(2) 서울시의 고등학교만 추출
학교급 구분 필드 값이 고등학교 이고, 소재지도로명주소에 서울이라는 단어가 포함된 경우

> SQL 의 LIKE 연산자를 str.contains() 로 사용할 수 있다

```python
df2 = df.query("학교급구분 == '고등학교' and 소재지도로명주소.str.contains('서울')" )
df2
```

#### 3-3 데이터 시각화

```python
#새로운 지도 객체 생성
map_osm5 = folium.Map(location = MY_PLACE , zoom_start= 12)


html =  "<font color='green' style='white-space: nowrap'><b>%s//b>//font>"

#데이터프레임의 행 수만큼 반복하면서 마커 생성
for i in df2.index:
  #행 우선 접근 방식으로 값 추출하기
  name = df2.loc[i,'학교명']
  lat = df2.loc[i,'위도']
  lng = df2.loc[i,'경도']


  # 추출한 정보를 지도에 표시
  popup_html = folium.Popup(html % name, parse_html=False)
  marker = folium.Marker([lat,lng], popup=popup_html)
  marker.add_to(map_osm5)

#웹 페이지 파일이 저장될 폴더 생성
if not path.exists('output'):
  mkdir('output')

#지도가 표시된 웹 페이지 파일이 저장된다
#결과 확인은 윈도우 폴더 창에서 직접 웹 페이지 파일 더블 클릭
map_osm5.save('output/map_osm5.html')

```
