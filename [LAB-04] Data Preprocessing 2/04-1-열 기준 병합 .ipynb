### #01. 준비작업

```python
from hossam import load_data
from pandas import DataFrame
from pandas import merge
```

```python
# 샘플 데이터 가져오기 - 고객 데이터와 매출 샘플 에이터
고객 = load_data('customer_no_name')
고객
```

###  #02.열 기준 병합 (1) 일치하는 데이터끼리 병합
- merge를 활용한 inner join
- 서로 일치하는 데이터끼리만 병합되고 ,일치 하지 않으면 버려진다
- key 값이 여러개 일치하면 경우의 수 만큼 행이 생성된다

```python
merge(고객,매출)
```

- SQL 의 LEFT OUTER JOIN

```python
merge(고객,매출,how='left')
```

###  #02.열 기준 병합 (3) 오른쪽 데이터 프레임을 기준으로 병합 
- SQL 의 RIGHT OUTER JOIN

```python
merge(고객,매출,how='right')
```

### #03. 병합 기준 열 지정하기 (1) on 을 활용한 기본 열 지정 후 병합

```python
#샘플 데이터 가져오기
cd1=load_data('customer_data1')
cd1
```

```python
cd2=load_data('customer_data2')
cd2
```

```python
#cd1 과 cd2 를 병합하게 되면, 이름이 같은 컬럼이 고객명과 데이터 2개 이므로 and 조건으로 조인을 실행하게 된다. 하지만 데이터 컬럼의 경우, cd1 은 int, cd2 는 str 이다 보니 데이터 타입이 달라 에러가 발생한다.

#이처럼 특정 컬럼에 대해서만 조인을 실행하려 할 떄는 병합하려는 열 기준을 지정해줘야한다 > on 사용

tmp = merge(cd1,cd2,on='고객명')
tmp
```

### #03. 병합 기준 열 지정하기 (2) 두 데이터 프레임의 모든 컬럼의 이름이 다른 경우
- df_left 의 이름 컬럼과 df_right 의 성함 컬럼을 기준으로 병합하고 싶을 떈, left_on ,right_on 옵션

```python
df_left = DataFrame({'이름':['영희','철수'],'국어':[87,91]})
df_left
```

```python
df_right = DataFrame({'성함':['영희','철수'],'영어':[90,82]})
df_right
```

```python
r3 = merge(df_left,df_right,left_on=['이름'],right_on=['성함'])
r3
```

### #04. 인덱스를 활용한 병합
- 컬럼이 아닌 인덱스를 통해서도 조인을 진행할 수 있다
- left_index=True 옵션으로 인덱스를 키값처럼 사용할 것인지 지정할 수 있다
- 디폴트는 inner join 이나 how 옵션으로 다른 조인도 가능

```python
df_left = DataFrame ({'수학':[90,82]},index =['민철','봉구'])
df_left
```

```python
df_right = DataFrame ({'국어':[90,82]},index =['민철','철수'])
df_right
```

```python
merge(df_left,df_right,left_index=True,right_index=True)
```

```python
merge(df_left,df_right,left_index=True,right_index=True,how='left')
```

```python
merge(df_left,df_right,left_index=True,right_index=True,how='right')
```

```python
merge(df_left,df_right,left_index=True,right_index=True,how='outer')
```

### #05. 인덱스와 컬럼을 각각 기준으로 하기
- 인덱스와 컬럼이름을 하나씩 사용하여 조인도 가능함

```python
df_left = DataFrame({'수학':[90,82]},index=['민철','봉구'])
df_left
```

```python
df_right = DataFrame({'성명':['민철','철수'],'영어':[90,82]})
df_right
```

```python
#왼쪽은 인덱스, 오른쪽은 컬럼을 기준으로 inner join
merge(df_left,df_right,left_index=True,right_on=['성명'])
```

```python
#왼쪽은 인덱스, 오른쪽은 컬럼을 기준으로 left outer join
merge(df_left,df_right,left_index=True,right_on=['성명'],how='left')
```

```python
#왼쪽은 인덱스, 오른쪽은 컬럼을 기준으로 right outer join
merge(df_left,df_right,left_index=True,right_on=['성명'],how='right')
```

```python
#왼쪽은 인덱스, 오른쪽은 컬럼을 기준으로 full outer join
tmp = merge(df_left,df_right,left_index=True,right_on=['성명'],how='outer')
tmp
```

```python
#병합 후에는 인덱스를 정리해야한다
tmp2 = tmp.set_index('성명')
tmp2
```
