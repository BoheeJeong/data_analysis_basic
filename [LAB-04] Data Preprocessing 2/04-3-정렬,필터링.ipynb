### #01.준비작업
- 라이브러리 참조 및 데이터 불러오기

```python
from hossam import load_data
```

```python
origin=load_data('grade')
origin
```

### #02.데이터 정렬 : 변수에 의한 정렬
- 기본값이 오름차순이기 때문에, 내림 차순은 ascending=False

```python
단일정렬df = origin.sort_values('학년',ascending=False)
단일정렬df
```

### #02.데이터 정렬 : 여러 변수 정렬
- 학년이 같은 경우 국어성적으로 정렬

```python

복수정렬df = origin.sort_values(['학년','국어'],
                            ascending=[False,False])

복수정렬df
```

### #02.데이터 정렬 : 인덱스를 기준으로 정렬
- 인덱스는 하나만 존재하므로 어떤 열로 정렬할 건지는 입력이 불필요

```python
인덱스정렬df =  origin.sort_index(ascending=False)
인덱스정렬df
```

### #03.필터링 : 행 단위 필터링 (조건 검색)

```python
#r1 = origin[origin['국어']>80] >> 동일 표현
r1 = origin.query('국어>80')  #조건을 " " 안에 넣어야한다
r1
```

```python
#r1 = origin[origin['성별'] =='남자'] >> 동일 표현
r2 = origin.query("성별 == '남자'")  #조건을 " " 안에 넣어야하고, str 값은 또 다른 따옴표로 감싸줘야한다
r2
```

```python
r3=origin.query('국어 > 80 and 수학 > 80')
r3
```

```python
r4=origin.query('국어 <= 80 or 수학 <=80')
r4
```

```python
#조건값을 변수로 적용하기 > @변수이름 형태로 조건식에 포함된다

point = int(input('점수를 입력하세요:'))

print('국어 점수가 %d 점 보다 높은 학생'%point)


r5=origin.query('국어 > @point')
r5
```

```python
#열 단위의 필터링
fl = origin.filter(['국어','영어'])
fl
```
