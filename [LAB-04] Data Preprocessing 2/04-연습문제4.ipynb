- 이 데이터를 분석하여 우리의 비즈니스에 도움이 될 핵심 인사이트를 발견하는 것이 당신의 업무입니다.

온라인 쇼핑몰의 다음 분기 프로모션 전략을 수립하기 위한 분석 업무입니다.

1. 데이터 정제

데이터에 결측치가 있는지 확인하고,Price 컬럼에 있는 결측치는 전체 상품의 평균 Price로 대체해주세요.

2. 베스트셀러 카테고리 분석

Category별로 주문 건수를 계산하여 가장 많은 주문이 발생한 카테고리 상위 3개를 찾아주세요.

3. 지역별 구매력 분석

Region별 평균 Price를 계산하여 어느 지역 고객이 평균적으로 가장 비싼 상품을 구매하는지 확인해주세요. 결과는 평균 구매액이 높은 순서대로 정렬하여 보여주세요.

```python
from hossam import load_data
import numpy as np
from pandas import DataFrame
from pandas import merge
from pandas import concat
from pandas import CategoricalDtype
from sklearn.impute import SimpleImputer
```

```python
shop_order_list = load_data('shop_order_list')
shop_order_list
```

```python
# 1. 데이터 정제
# 데이터에 결측치가 있는지 확인하고,Price 컬럼에 있는 결측치는 전체 상품의 평균 Price로 대체해주세요.

# price 컬럼의 평균값을 구한다
shop_order_list_copy = shop_order_list.copy()
shop_order_list_mean=shop_order_list_copy['Price'].mean()


#구한 평균값을 기존 테이블에 넣어준다
col=shop_order_list_copy['Price'].fillna(shop_order_list_mean)
shop_order_list_copy['Price']=col
shop_order_list_copy
```

```python
# 2. 베스트셀러 카테고리 분석

# Category별로 주문 건수를 계산하여 가장 많은 주문이 발생한 카테고리 상위 3개를 찾아주세요.

#구하려는 데이터 컬럼만 필터링 작업
shop_order_list_copy2=shop_order_list_copy.copy()

category_top3 = shop_order_list_copy2.filter(['Category','UserID'])
category_top3
```

```python
a=category_top3.groupby('Category').count()
a.sort_values('UserID',ascending=False).head(3)
```

```python
# 3. 지역별 구매력 분석

# Region별 평균 Price를 계산하여 어느 지역 고객이 평균적으로 가장 비싼 상품을 구매하는지 확인해주세요. 결과는 평균 구매액이 높은 순서대로 정렬하여 보여주세요.

#Region별 평균 Price

region_mean_sorted = shop_order_list_copy2.filter(['Region','Price'])
region_mean_sorted
```

```python
avg_price=region_mean_sorted.groupby('Region').mean()
avg_price.sort_values('Price',ascending=False)

```
