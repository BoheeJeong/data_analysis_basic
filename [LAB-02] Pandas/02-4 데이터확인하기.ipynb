### 특정 행에 접근하기

### 단일 값 확인하기

```python
#1번쨰 행의 2번쨰 열 데이터
origin.iloc[1,2]
```

```python
#모든 행에서 3번쨰 열만 가져옴
origin.iloc[:,3]
```

### 값의 변경

### 기초 통계량 확인
- 데이터 프레임은 기본적으로 시리즈 객체이므로 기술 통계량 관련 메서드를 모두 사용할 수 있음

```python
print("합계:", origin['국어'].sum())
print("평균:", origin['국어'].mean())
print("최대:", origin['국어'].max())
print("최소:", origin['국어'].min())
print("분산:", origin['국어'].var())
print("표준편차:", origin['국어'].std())
print("중앙값:", origin['국어'].median())
print("1사분위 수:", origin['국어'].quantile(0.25))
print("2사분위 수:", origin['국어'].quantile(0.5))
print("3사분위 수:", origin['국어'].quantile(0.75))
```
