### Series 객체 생성(1)
- list 와 ndarrat 를 통해 생성할 수 있다
- Series는 기존의 연속형 자료형을 가공하여 생성된 데이터 구조

```python
#리스트를 통한 객체 생성
my_list =[10,30]
s1 = Series(my_list)
s1
```

```python
#ndarray 를 통한 객체 생성
my_array = np.array([100,200])
s2=Series(my_array)
s2
```

### Series 객체의 인덱싱과 슬라이싱
- list 나 ndarray 를 통해 생성한 경우
- 일반적인 연속성 자료형의 경우와 동일하게 슬라이싱이 가능하다

```python
my_series = Series([10,20,30,40,50])
print("===인덱싱===")
print(my_series[0])

print("===슬라이싱===")
print(my_series[1:3])   # 1번쨰와 2번쨰만 슬라이싱
```

딕셔너리를 통해 생성한 경우, 인덱스가 문자열로 존재한다

```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})
print("===인덱싱===")
print(my_series["a"])

print("===슬라이싱===")
print(my_series["b":"d"])   # b,c,d 슬라이싱 (즉, 숫자형과 다른 규칙 적용)
```

### Series 의 값만 추출하기
- values 프로퍼티를 이용하여 값만 추출할 수 있다

```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})


print(type(my_series.values))
my_series.values
```

- 값을 list 자료형으로 변환

```python
mylist = list(my_series.values)
mylist
```

### Series 의 인덱스만 추출하기
- index 프로퍼티를 사용하여 값만 추출할 수 있다
- ndarray 형식



```python
my_series = Series({"a":100,"b":200,"c":300,"d":400})


print(type(my_series.index))
my_series.index
```

인덱스를 list 자료형으로 변환

```python
mylist = list(my_series.index)
mylist
```
