## 02 Series 객체의 기능 | 연습문제
- 다음은 어떤 학생의 5과목에 대한 시험 점수를 딕셔너리로 나타낸 것 입니다.
- scores = {"국어": 85, "영어": 92, "수학": 78, "과학": 95, "사회": 88}
- Pandas의 `Series`를 사용하여 다음 작업을 수행하는 코드를 작성하세요

 1. 위 딕셔너리를 `Series` 객체로 변환하세요.
 2. 수학 과목의 점수를 출력하세요.
 3. 점수가 90점 이상인 과목만 선택하여 출력하세요.
 4. 모든 과목 점수의 평균을 계산하여 출력하세요.

```python
import numpy as np
from pandas import Series

# 1. 위 딕셔너리를 `Series` 객체로 변환하세요.
scores = Series({"국어": 85, "영어": 92, "수학": 78, "과학": 95, "사회": 88})


# 2. 수학 과목의 점수를 출력하세요.
print("수학 점수는",scores["수학"])
```

```python
# 3. 점수가 90점 이상인 과목만 선택하여 출력하세요.
print(scores[scores>=90])
```

```python
# 4. 모든 과목 점수의 평균을 계산하여 출력하세요.
print(scores.mean())
```
